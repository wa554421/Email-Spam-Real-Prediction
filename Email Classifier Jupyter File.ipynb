```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
call = pd.read_csv("spam_ham_dataset.csv")
```

```python
call["label"].value_counts()
```

```python
call
```

```python
call["label"] = call["label"].map({"ham":1,"spam":0})
```

```python
call
```

```python
call["text"] = call["text"].str.lower() # in pandas 
```

```python
call
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
# Convert text into numbers
```

```python
tfidf = TfidfVectorizer(stop_words="english",max_features=3000) # CONVERT TEXT INTO NUMERICAL FORM THAT MACHINE LEARNING CAN UNDERSTOOD
```

```python
x = tfidf.fit_transform(call["text"])
```

```python
y = call["label"]
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)# testsize means 20% data for testing and
# and random state is fixed number that make sure the random split(train and testing) is always same whenever yiu run the code
# wihtour random state you run code and always different split data in train and testing
```

```python
from sklearn.naive_bayes import MultinomialNB # best classifier model for text Classification after tfidfvectorizer
```

```python
model = MultinomialNB()
```

```python
model.fit(x_train,y_train)
```

```python
pred = model.predict(x_test[0])
```

```python
pred
```

```python
sample_email = ["tomorrow you can contact me for buisness deals"]
sample_tfidf = tfidf.transform(sample_email) # simply transform new text email into tfidf feature  that ysed previous for training
```

```python
prediction = model.predict(sample_tfidf) # predict this user giver email
print("Spam" if prediction == 0 else "Real") 
```

```python
import joblib # Saving Model

joblib.dump(model, 'spam_classifier_model.pkl')

# Save vectorizer
joblib.dump(tfidf, 'tfidf_vectorizer.pkl')
#It stores how it learned to convert words into numbers means prevous trainig data features(vocabulary, IDF weights, etc.)
# You must use the same vectorizer that was trained before — otherwise, the model won’t understand the input format.
```
